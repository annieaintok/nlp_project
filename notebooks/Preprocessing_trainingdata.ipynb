## Preparation

```python
# creating list with all pdf file names
pdf_names = []
for x in range(0,7):
  cat = ['asset_prcing_', 'banking_', 'corporate_finance_', 'corporate_governance_', 'insurance_', 'investment_and_portfolio_management_', 'market_microstructure_']
  for i in range(0, 5):
      name = f'{cat[x]}{i+1}.pdf'
      pdf_names.append(name)
```

## Importing text from pdf files with PyMuPDF



```python
def pdf_to_string(filename):
    with fitz.open(f"{filename}") as doc:
        text1 = ""
        for page in doc:
            text1 += page.get_text()
        return text1
```

```python
liste = []

#extract text of all pdfs
for x in pdf_names:
#    full_text = []
    ans=pdf_to_string(x)
#make list of many strings to one long string
    #ans = ' '
    #for i in full_text:
      # concatenating the strings
      # using + operator
        #ans = ans+ ' '+ i
#remove braces, make lowercase, strip whitespace, replace, 
    opening_braces = '\(\['
    closing_braces = '\)\]'
    non_greedy_wildcard = '.*?'
    ans = re.sub(f'[{opening_braces}]{non_greedy_wildcard}[{closing_braces}]', ' ', ans)
    ans = ans.lower()
    sub_str = "references"
    res = ans.split(sub_str)#[:-1]
    fin_text = ' '
    for i in res: 
        fin_text = fin_text+ ' '+ i
    #print(f"----------------------ARTICLE {x}-----------------", res[0])
    fin_text = fin_text.strip()
    fin_text = fin_text.replace('\n', ' ')
    fin_text = fin_text.replace('  ', ' ')
    liste.append(fin_text)
```
